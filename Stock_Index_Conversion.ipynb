***Dropping the NAN columns***

```python
df=pd.DataFrame(iD)
df.dropna()
df
```

```python
df[['Index',"Open",'High',"Low",'Close','Adj Close']]
```

```python
df
```

**(0) United States	New York Stock Exchange	NYA	USD**

1.00 US Dollar =74.243277 Indian Rupees


```python
f=df[df.Index=='NYA']
# dropping the rows having NaN values
df = df.dropna()
# To reset the indices 
df = df.reset_index(drop = True)
df['INR_close']=df['Adj Close']*74.243277
print(df[df.Index=='NYA'])
```

```python
figure(figsize=(16, 8), dpi=80)
plt.plot(f['Date'],f['Close'])
```
